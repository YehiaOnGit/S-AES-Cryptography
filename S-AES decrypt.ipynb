# rev Round 2

```python
#round2_outpt = ciphertext xor k2
round2_outpt=int(ciphertxt,2) ^ int(k2,2)
print(round2_outpt)
round2_outpt=format(round2_outpt,'016b')
print(round2_outpt)

#inverse shift row
frst = round2_outpt[0:4] # extract first 4 bits
scnd = round2_outpt[4:8] # extract second 4 bits
thrd = round2_outpt[8:12] # extract third 4 bits
frth = round2_outpt[12:16] # extract fourth 4 bits


scnd, frth = frth, scnd

subnb_befor=frst+scnd+thrd+frth
#subnib rwnd 2 look in 1,3 cols
frst = subnb_befor[0:4] # extract first 4 bits
scnd = subnb_befor[4:8] # extract second 4 bits
thrd = subnb_befor[8:12] # extract third 4 bits
frth = subnb_befor[12:16] # extract fourth 4 bits


for row in range(len(subnibTBL)):
    for col in [1, 3]:
        if subnibTBL[row][col] == frst:
            sr=row
            sc=col
            sub1=subnibTBL[sr][sc-1] 
            print(f"Found part1 {sub1} ")  
        if subnibTBL[row][col] == scnd:
            sr=row
            sc=col
            sub2=subnibTBL[sr][sc-1] 
            print(f"Found part2 {sub2} ")
        if subnibTBL[row][col] == thrd:
            sr=row
            sc=col
            sub3=subnibTBL[sr][sc-1] 
            print(f"Found part3 {sub3} ")
        if subnibTBL[row][col] == frth:
            sr=row
            sc=col
            sub4=subnibTBL[sr][sc-1] 
            print(f"Found part4 {sub4} ")

input_rwnd2=sub1 + sub2 +sub3 +sub4  
print(input_rwnd2)
######################################  input rwnd 2
```
